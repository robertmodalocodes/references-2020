## Functional plot

```python
x_1 = np.linspace(0, 5, 10)
y_1 = x_1 ** 2
plt.plot(x_1, y_1)
plt.title('Squared floats chart')
plt.xlabel('Floats')
plt.ylabel('Floats squared')
# plt.show() . use this code at the end of the plt lines if not using jupyter notebook
```

## Print multiple plots

```python
plt.subplot(1, 2, 1) # n row, n column, chart 1
plt.plot(x_1, y_1, 'r')
plt.subplot(1, 2, 2) # n row, n column, chart 2
plt.plot(x_1, y_1, 'b')
```

## Using figure objects

```python
fig_1 = plt.figure(figsize=(5, 4), dpi=300)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Days squared chart')
axes_1.plot(x_1, y_1, label='x/$x^{2}$')
axes_1.plot(y_1, x_1, label='$x^{2}$/x')
# legend positions
# upper right: 1, upper left: 2, lower left: 3, lower right: 4, default: 0
# or supply a tuple of x & y from lower left
axes_1.legend(loc=1)

# add/insert another chart in a figure
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3]) # notice that we're still using the same figure (fig_1)
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days squared')
axes_2.set_title('Days squared chart')
axes_2.plot(x_1, y_1, 'r')
axes_2.plot(y_1, x_1, 'g')
axes_2.plot(x_1 ** 2 * 2, y_1 ** 2)
axes_2.plot(y_1 ** 2, x_1 ** 2 * 2)

axes_2.text(100, 100, 'Sample text') # these position values (coordinates) depends on your chart
```

## Pie charts

```python
import random
    
fig_6 = plt.figure(figsize=(8, 5), dpi=100)
axes_6 = fig_6.add_axes([0.1, 0.1, 0.9, 0.9])

# create a pie chart of the number of pokemon by type
types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire',
        'Poison', 'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric',
        'Dragon', 'Fairy', 'Ghost', 'Ice']

poke_num = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 40]

# create random numbers to set rgb
colors = []
for i in range(18):
    rgb = (random.uniform(0, 0.5), random.uniform(0, 0.5), random.uniform(0, 0.5))
    colors.append(rgb)

explode = [0] * 18
explode[0] = 0.5

wedges, texts, autotexts = plt.pie(poke_num, explode=explode, labels=types,
                                  colors=colors, autopct='%1.0F%%', shadow=True,
                                  startangle=140, textprops=dict(color='w'))

plt.legend(wedges, types, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))

```

## 3D surface

```python
# scatter
from mpl_toolkits import mplot3d

fig_9 = plt.figure(figsize=(8, 5), dpi=100)
axes_9 = fig_9.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')
# always add "projection='3d' for 3d plots"

z_3 = 15 * np.random.random(100)
x_3 = np.sin(z_3) * np.random.randn(100)
y_3 = np.cos(z_3) * np.random.randn(100)
axes_9.scatter3D(x_3, y_3, z_3, c=z_3, cmap='Blues')
```

```python
# contour plot

# create a function do determine z values from x and y values
# always do this first for creating a contour plot
fig_10 = plt.figure(figsize=(8, 5), dpi=100)
axes_10 = fig_10.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')

def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x_4 = np.linspace(-6, 6, 30)
y_4 = np.linspace(-6, 6, 30)

x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4, y_4)

axes_10.view_init(45, 55)
axes_10.set_xlabel('x')
axes_10.set_ylabel('y')
axes_10.set_zlabel('z')
axes_10.contour3D(x_4, y_4, z_4, 20, cmap='Blues')
```

```python
# 3d wireframe
fig_11 = plt.figure(figsize=(8, 5), dpi=100)
axes_11 = fig_11.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')

def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x_5 = np.linspace(-6, 6, 30)
y_5 = np.linspace(-6, 6, 30)

x_5, y_5 = np.meshgrid(x_5, y_5)
z_5 = get_z(x_5, y_5)

axes_11.view_init(45, 55)
axes_11.set_xlabel('x')
axes_11.set_ylabel('y')
axes_11.set_zlabel('z')
axes_11.plot_wireframe(x_5, y_5, z_5, color='blue')
```

```python
# 3d surface
fig_12 = plt.figure(figsize=(8, 5), dpi=100)
axes_12 = fig_12.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')

def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x_6 = np.linspace(-6, 6, 30)
y_6 = np.linspace(-6, 6, 30)

x_6, y_6 = np.meshgrid(x_6, y_6)
z_6 = get_z(x_6, y_6)

axes_12.view_init(45, 55)
axes_12.set_xlabel('x')
axes_12.set_ylabel('y')
axes_12.set_zlabel('z')
axes_12.plot_surface(x_6, y_6, z_6, rstride=1, cstride=1, cmap='Blues', edgecolor='none')
```

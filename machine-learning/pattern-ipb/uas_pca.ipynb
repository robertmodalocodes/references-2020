```python
data = pd.read_csv("D:/Datasets/HCV-Egy-Data/HCV-Egy-Data.csv")
data.head(4)
```

```python
# x = data.drop(columns=["Baselinehistological staging"])
x = data
x.head(4)
```

```python
# normalisasi
x_norm = StandardScaler().fit_transform(x)
x_norm.shape
```

```python
# cek data ternormalisasi
print(np.mean(x_norm), np.std(x_norm))
x_norm
```

```python
# konversi array data ternormalisasi ke bentuk tabel
fitur = x.columns
fitur_array = np.array(fitur)
fitur_array
```

```python
# konversi array data ternormalisasi ke bentuk tabel
fitur_df = pd.DataFrame(x_norm, columns=fitur_array)
fitur_kolom = ['fitur'+str(i) for i in range(x_norm.shape[1])]
fitur_norm = pd.DataFrame(x_norm, columns=fitur_kolom)
fitur_norm.tail(4)
```

```python
# mulai PCA dengan memproyeksi data
# 29 kolom/dimensi menjadi 2 dimensi
from sklearn.decomposition import PCA

pca_hcv = PCA(n_components=2)
principalComponents_hcv = pca_hcv.fit_transform(fitur_norm)
```

```python
principal_hcv_df = pd.DataFrame(data = principalComponents_hcv, columns=['principal component 1', 'principal component 2'])
principal_hcv_df.tail(4)
```

```python
# varian setelah proyeksi data
print(f'Explained variation per principal component: {pca_hcv.explained_variance_ratio_}')
```

```python
print (principal_hcv_df.loc)   
print (principal_hcv_df.loc[40:60])    
principal_hcv_df.describe()
```

```python
classes = ['None', 'Portal Fibrosis', 'Few Septa', 'Many Septa', 'Cirrhosis']

y_ = x['Baselinehistological staging']

principal_hcv_df_label = principal_hcv_df
principal_hcv_df_label['y'] = y_

principal_hcv_df_label.y[principal_hcv_df_label.y == 1] = classes[1]
principal_hcv_df_label.y[principal_hcv_df_label.y == 2] = classes[2]
principal_hcv_df_label.y[principal_hcv_df_label.y == 3] = classes[3]
principal_hcv_df_label.y[principal_hcv_df_label.y == 4] = classes[4]

print(principal_hcv_df.head(4))
principal_hcv_df_label.head(4)

```

```python
import seaborn as sns
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="principal component 1", y="principal component 2",
    hue="y",
    palette=sns.color_palette("hls", 4),
    data=principal_hcv_df_label,
    legend="full",
    alpha=0.3
)
```

```python
# analisis data ternormalisasi dengan LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(fitur_norm, y_) 

lda_hcv = LinearDiscriminantAnalysis(n_components=2)
# prepare transform on dataset
linearDiscriminant_hcv = lda_hcv.fit(x_train, y_train).transform(x_train)

```

```python
ld_hcv_df = pd.DataFrame(data = linearDiscriminant_hcv, columns=['LD 1'])
ld_hcv_df.tail(4)
```

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

plt.show()
```
